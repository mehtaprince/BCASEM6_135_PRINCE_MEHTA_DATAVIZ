```python
import pandas as pd

# Load dataset
dataset = pd.read_csv("student_data.csv")

# Mean
print("Mean:\n", dataset.mean(numeric_only=True))

# Median
print("\nMedian:\n", dataset.median(numeric_only=True))

# Mode
print("\nMode:\n", dataset.mode(numeric_only=True))

# Range
range_values = dataset.max(numeric_only=True) - dataset.min(numeric_only=True)
print("\nRange:\n", range_values)

# Variance
print("\nVariance:\n", dataset.var(numeric_only=True))

# Quartiles
print("\nQuartiles:\n", dataset.quantile([0.25, 0.5, 0.75], numeric_only=True))
```
